```r
# a. Build the frequency distribution table and probability distribution table. Check that the distribution sum of the probability is 1. Plot the probability distribution table

# 1. Define the data
X <- c(9,7,8,9,8,7,8,11,8,8,6,8,8,11,11,9,9,8,9,6,10,9,6,4,8,8,9,6,8,9,10,8,8,9,6,9,8,9,6,10,10,8,5,8,11,10,9,7,7,6,7,8,9,7,8,8,10,8,7,6,7,8,10,10,7,9,8,9,8,9,7,4,7,5,8,9,7,10,6,9,8,8,8,8,8,8,9,7,8,8,11,9,11,8,9,7,9,9,8,8,5,6,7,6,8,10,10,8,6,6,8,7,7,8,8,8,6,7,8,8,8,9,10,8,5,9,8,8,6,6,6,8,8,7,7,10,10,7,9,7,6,9,8,5,8,7,10,5,5,8,6,10,8,7,9,6,8,8,7,10,10,10,10,6,8,8,6,10,8,10,7,7,10,7,8,7,10,10,7,7,8,7,10,8,8,7,7,7,8,8,5,8,5,7,8,8,8,10,9,10)
```

```r
# Calculate frequencies of each unique value in X
frequency_table <- table(X)

# Convert to a data frame for better presentation
frequency_df <- as.data.frame(frequency_table)
colnames(frequency_df) <- c("Defects", "Frequency")

cat("--- Frequency Distribution Table ---\n")
print(frequency_df)
cat("\n")
```

```r
# Build the probability distribution table

# Calculate the total number of observations
total_observations <- length(X)

# Calculate probabilities
probability_distribution <- frequency_table / total_observations

# Convert to a data frame for better presentation
probability_df <- as.data.frame(probability_distribution)
colnames(probability_df) <- c("Defects", "Probability")

cat("--- Probability Distribution Table ---\n")
print(probability_df)
cat("\n")
```

```r
# Check that the distribution sum of the probability is 1
sum_of_probabilities <- sum(probability_distribution)
cat("Sum of Probabilities:", sum_of_probabilities, "\n")
if (abs(sum_of_probabilities - 1) < 1e-9) { # Use a small tolerance for floating point comparison
  cat("The sum of probabilities is indeed 1.\n\n")
} else {
  cat("Warning: The sum of probabilities is not exactly 1. There might be a rounding issue or calculation error.\n\n")
}
```

```r
# Plot the probability distribution table

# Ensure 'Defects' column in probability_df is numeric for proper plotting
probability_df$Defects <- as.numeric(as.character(probability_df$Defects))

# Create a bar plot of the probability distribution
ggplot(probability_df, aes(x = Defects, y = Probability)) +
  geom_bar(stat = "identity", fill = "skyblue", color = "black") +
  labs(title = "Probability Distribution of Defects per 1000 Parts",
       x = "Number of Defects (X)",
       y = "Probability (P(X))") +
  theme_minimal() +
  scale_x_continuous(breaks = unique(probability_df$Defects)) + # Ensure all defect values are shown on x-axis
  geom_text(aes(label = round(Probability, 3)), vjust = -0.5, size = 3) # Add probability values on bars
```

<b> Observation: </b> 
- Frequency and Probability Distribution: This step organizes raw defect data into tables and a visual plot, clearly showing the observed likelihood of each defect count and confirming that all probabilities sum to one.

```r
# b. Determine the expected value (mean), variance, and standard deviation
cat("--- Descriptive Statistics ---\n")

# Expected Value (Mean)
expected_value <- sum(probability_df$Defects * probability_df$Probability)
cat("Expected Value (Mean):", round(expected_value, 3), "\n")
```

```r
# Variance
# Variance = E[X^2] - (E[X])^2
# Calculate E[X^2]
expected_value_x_squared <- sum((probability_df$Defects^2) * probability_df$Probability)
variance <- expected_value_x_squared - (expected_value^2)
cat("Variance:", round(variance, 3), "\n")
```

```r
# Standard Deviation
standard_deviation <- sqrt(variance)
cat("Standard Deviation:", round(standard_deviation, 3), "\n")
```

<b> Observation: </b> 
- Expected Value, Variance, and Standard Deviation: These calculations provide the average number of defects expected, along with measures of how much the actual defect counts typically vary from that average.

```r
# c. Determine each probability listed
cat("\n--- Probabilities ---\n")

# (i) P(6 <= X <= 9)
# Filter the probability_df for defects between 6 and 9 inclusive
prob_6_to_9 <- sum(probability_df$Probability[probability_df$Defects >= 6 & probability_df$Defects <= 9])
cat("P(6 <= X <= 9):", round(prob_6_to_9, 3), "\n")
```

```r
# (ii) P(X >= 9)
# Filter the probability_df for defects greater than or equal to 9
prob_greater_than_or_equal_to_9 <- sum(probability_df$Probability[probability_df$Defects >= 9])
cat("P(X >= 9):", round(prob_greater_than_or_equal_to_9, 3), "\n")
```

<b> Observation: </b> 
- Specific Probabilities: These calculations determine the likelihood of observing defect counts within a specific range

2) Data documented for weather events in the state of Montana is obtained, over a period of years, and includes hail storms causing property damage. The data is analyzed and appears to fit a Poisson distribution with average of 3.3 damaging hail storms per year. Answer the following to 3 significant digits.

a. What is the likelihood that between 2 and 4 damaging hail storms are seen in a randomly selected year i.e. P(2 ≤ X ≤ 4) given µ=3.3. 
b. What is the probability that at least one significant hail storm causing damage is experienced in the coming year, P(X ≥ 1) given µ=3.3
c. What is the likelihood that Montana would experience between 5 and 7 damaging hailstorms over a two-year period: P(5 ≤ X ≤ 7) given µ=6.6

```r
# a. What is the likelihood that between 2 and 4 damaging hail storms are seen in a randomly selected year, i.e., P(2 ≤ X ≤ 4) given µ=3.3.

# Given parameters for the Poisson distribution
mu <- 3.3

# Calculate P(2 <= X <= 4)
# This is equivalent to P(X=2) + P(X=3) + P(X=4)
# Using dpois for individual probabilities:
prob_x_eq_2 <- dpois(2, lambda = mu)
prob_x_eq_3 <- dpois(3, lambda = mu)
prob_x_eq_4 <- dpois(4, lambda = mu)

prob_2_to_4_sum <- prob_x_eq_2 + prob_x_eq_3 + prob_x_eq_4

cat("The likelihood that between 2 and 4 damaging hail storms are seen (P(2 <= X <= 4)) is:\n")
cat("Using sum of dpois:", round(prob_2_to_4_sum, 3), "\n")
```

```r
# Alternatively, using ppois for cumulative probabilities:
# P(2 <= X <= 4) = P(X <= 4) - P(X <= 1)
prob_2_to_4_cumulative <- ppois(4, lambda = mu) - ppois(1, lambda = mu)

cat("The likelihood that between 2 and 4 damaging hail storms are seen (P(2 <= X <= 4)) is:\n")
cat("Using ppois (cumulative):", round(prob_2_to_4_cumulative, 3), "\n")
```

<b> Observation: </b> 
- Likelihood of 2 to 4 Storms: This calculation determines the specific probability of observing a moderate range of damaging hail storms within a single year, directly from the Poisson distribution = 60.4%

```r
# b. What is the probability that at least one significant hailstorm causing damage is experienced in the coming year, P(X ≥ 1) given µ=3.3

# Given parameters for the Poisson distribution
mu <- 3.3

# Calculate P(X = 0) using dpois (probability mass function)
prob_x_eq_0 <- dpois(0, lambda = mu)

# Calculate P(X >= 1) using the complement rule
prob_at_least_1 <- 1 - prob_x_eq_0

cat("The probability that at least one significant hail storm is experienced (P(X >= 1)) is:\n")
cat("P(X >= 1) =", round(prob_at_least_1, 3), "\n")
```

<b> Observation: </b> 
- Probability of At Least One Storm: This calculation uses the complement rule to efficiently find the high likelihood of experiencing at least one damaging hailstorm in a given year = 96.3%

```r
# c. What is the likelihood that Montana would experience between 5 and 7 damaging hailstorms over a two-year period: P(5 ≤ X ≤ 7) given µ=6.6

# Mean for a two-year period
mu_two_years <- 6.6 # Given directly as 6.6, which is 2 * 3.3

# P(5 <= X <= 7) = P(X <= 7) - P(X <= 4)
prob_x_le_7_two_years <- ppois(7, lambda = mu_two_years)
prob_x_le_4_two_years <- ppois(4, lambda = mu_two_years)

prob_5_to_7_two_years <- prob_x_le_7_two_years - prob_x_le_4_two_years

cat("Likelihood of between 5 and 7 damaging hail storms over a two-year period (P(5 <= X <= 7)):\n")
cat("P(5 <= X <= 7) =", format(round(prob_5_to_7_two_years, 3), nsmall = 3), "\n")
```

<b> Observation: </b> 
- Likelihood over Two Years: This calculation extends the Poisson model to a longer period, determining the probability of a specific range of hail storms by adjusting the mean for the extended timeframe 44.5%

3) A company that manufactures electronic switches tests each lot before shipping by taking a sample of 50 parts. If 2 or more are deemed defective, the lot is pulled for further inspection.

a.If the shipment actually contains 8.3% defective parts, what is the probability that 2 or more fail the test?  Indicate the answer to 4 significant figures.P(X>=2). (Note: remember that with pbinom, if lower.tail is false, use x-1 instead of x)
b. Calculate the mean and standard deviation of the number of parts that will pass the test for a sample of 50 from a shipment with the above proportion of defective parts. Answer to one decimal place. (Note: Question is for mean number that pass, not those that fail)

```r
# a. If the shipment actually contains 8.3% defective parts, what is the probability that 2 or more fail the test?  Indicate the answer to 4 significant figures.P(X>=2). 
# (Note: remember that with pbinom, if lower.tail is false, use x-1 instead of x)

# Given parameters for the binomial distribution
n <- 50    # Sample size (number of trials)
p <- 0.083 # Probability of a defective part (success)

# We want to find P(X >= 2)
# Using the complement rule: P(X >= 2) = 1 - P(X < 2) = 1 - P(X <= 1)
# Using pbinom with lower.tail = FALSE: P(X >= k) = pbinom(k - 1, size, prob, lower.tail = FALSE)
# So, P(X >= 2) = pbinom(2 - 1, size = n, prob = p, lower.tail = FALSE)
prob_x_ge_2 <- pbinom(1, size = n, prob = p, lower.tail = FALSE)

cat("Probability that 2 or more parts fail the test (P(X >= 2)):\n")
cat("P(X >= 2) =", format(round(prob_x_ge_2, 4), nsmall = 4), "\n") # Format to 4 decimal places
```

<b> Observation: </b> 
- This calculation determines the probability that 2 or more electronic switches in a sample of 50 will be defective, using the binomial distribution to assess lot quality = 92.74%

```r
# b. Calculate the mean and standard deviation of the number of parts that will pass the test for a sample of 50 from a shipment with the above proportion of defective parts. Answer to one decimal place.
# (Note: Question is for mean number that pass, not those that fail)

# Probability of a part passing the test
q <- 1 - p

# Mean (Expected Value) of parts that pass
mean_pass <- n * q

# Standard Deviation of parts that pass
# Variance of parts that pass = n * q * p
variance_pass <- n * q * p
sd_pass <- sqrt(variance_pass)

cat("\n--- Statistics for Parts Passing the Test ---\n")
cat("Mean number of parts that pass:", round(mean_pass, 1), "\n")
cat("Standard deviation of parts that pass:", round(sd_pass, 1), "\n")
```
